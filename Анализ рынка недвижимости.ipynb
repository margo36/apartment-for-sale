####  Рассмотрю типы данных в каждом столбце, найду столбцы, в которых нужно изменить тип данных и преобразую их.

```python
df.info()
# Вывела на экран основную информацию о датафрейме и изучаю тип данных
```

```python
# По результату изучения типы данных столбцов произвожу следующие изменения:

# Столбец `last_price` преобразую в int
df['last_price'] = df['last_price'].round()
df['last_price'] = df['last_price'].astype('int')

# Столбец `first_day_exposition` преобразую в datatime
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d %H:%M:%S')

# Столбец `floors_total` преобразую в int
df['floors_total'] = df['floors_total'].astype('int')

# Столбец `balcony` преобразую в int
df['balcony'] = df['balcony'].astype('int')
```

**Причины изменения тип данных:**  
    
Все цельночисленные данные, такие как цена, этажность дома, количество балконов лучше хранить в типе данных int, чтою не получить ошибку или неожиданный результат. Вещественными числами эти показатели быть не могут. 
    
Для корректной работы с датами необходим тип datetime. Этот тип позволяет управлять датами и временем, а также представлять их в удобном виде.

```python
sorted(df['locality_name'].unique())
# Вывожу на экран уникальные значения в столбце 'locality_name'
```

```python
# Заменяю все "ё" на "е", чтобы все "посёлки" стали "поселками" и исчезли неявные дубликаты названий населенных пунктов.
df['locality_name'] = df['locality_name'].str.replace('ё', 'е', regex=True)

# Заменяю одинаковые названия по статусу населенных пунтов.
df['locality_name'] = df['locality_name'].str.replace('поселок при железнодорожной станции', 'поселок станции', regex=True)
df['locality_name'] = df['locality_name'].str.replace('городской поселок', 'поселок городского типа', regex=True)

# Разные названия одного населенного пункта меняю на одинаковое.
df['locality_name'] = df['locality_name'].replace('поселок Мурино', 'Мурино')
df['locality_name'] = df['locality_name'].replace('деревня Кудрово', 'Кудрово')
df['locality_name'] = df['locality_name'].replace('поселок Рябово', 'поселок городского типа Рябово')
df['locality_name'] = df['locality_name'].replace('поселок станции Громово', 'поселок Громово')

# Исключение: коттеджный поселок Лесное находится в Ломоносовском районе, а поселок Лесное во Всеволожском.
```

Исключение: `коттеджный поселок Лесное` находится в Ломоносовском районе, а `поселок Лесное` во Всеволожском. Это разные населённые пункты.

```python
# Вывожу на экран уникальные значения в столбце 'ceiling_height'
sorted(df['ceiling_height'].unique())
```

**Из результата вызова метода `unique` можно сделать некоторые выводы:**

- а значения 25 м или 32 м скорее всего неправильно введены и логично предположить, что на самом деле это вещественные значения: 2.5 м и 3.2 м.
    
В своде правил 54.13330.2016 "СНиП 31-01-2003 Здания жилые многоквартирные" установлено, что в многоквартирных домах в северо-западной части России потолок должен быть не ниже 2,5 м.
Квартиры, у которых высота потолков превышает 4 метра, чаще всего относятся к классу "люкс" и стоят очень дорого. Однако, не стоит забывать, что в центре Санкт-Петербурга, в старом фонде высота потолков может достигать 6 и более метров.

```python
# Посмотрю, что из себя представляет квартира с высотой потолка 100 м
df[df['ceiling_height'] == 100]
```

Странная квартира: однокомнатная, 5 балконов, потолки 100 метров, общая площадь 25 кв.м. и цена 15 млн.. Считаю, что это вброс и её надо удалить

```python
df=df[df['ceiling_height'] !=100]
```

```python
# Обработаю значения, которые были введены неправильно - высота потолков 24.0, 25.0, 26.0, 27.0, 27.5 и 32.0.
df['ceiling_height'] = df['ceiling_height'].replace(24.0, 2.4)
df['ceiling_height'] = df['ceiling_height'].replace(25.0, 2.5)
df['ceiling_height'] = df['ceiling_height'].replace(26.0, 2.6)
df['ceiling_height'] = df['ceiling_height'].replace(27.0, 2.7)
df['ceiling_height'] = df['ceiling_height'].replace(27.5, 2.7)
df['ceiling_height'] = df['ceiling_height'].replace(32.0, 3.2)
```

```python
# Другие высокие показатели лучше удалить, так как природа аномалии непонятна.
df=df[df['ceiling_height'] !=8]
df=df[df['ceiling_height'] !=8.3]
df=df[df['ceiling_height'] !=10.3]
df=df[df['ceiling_height'] !=14]
df=df[df['ceiling_height'] !=20]
df=df[df['ceiling_height'] !=22.6]

#Так же удалю обьекты,у которых высота потолков менее 2.3 (20 см от нормы оставлю для натежных потолков).
df=df[df['ceiling_height'] !=1]
df=df[df['ceiling_height'] !=1.2]
df=df[df['ceiling_height'] !=1.75]
df=df[df['ceiling_height'] !=2]
df=df[df['ceiling_height'] !=2.2]
df=df[df['ceiling_height'] !=2.25]
```

```python
# Последним этапом обработаю пропуски, заполню медианой.
df['ceiling_height'].fillna(df['ceiling_height'].median(), inplace=True)
```

**Вывод:** Была удалена квартира с потолками 100 м, а так же объекты с высотой потолков до 2.3 м и свыше 8 м. А неверно заполненные значения (24.0, 25.0, 26.0, 27.0, 27.5 и 32.0 м) были приведены в порядок.

**Комментарий: Возможно, у домов одинаковой этажности высота потолков одинаковая. Но, хрущевки, брежневки и сталинки 5-этажные. Однако, высота потолков у них разная.**

### Добавлю в таблицу новые столбцы

```python
# Добавляю новый столбец с ценой квадратного метра.
df['price_m2'] = df['last_price'] / df['total_area']
df['price_m2'] = df['price_m2'].round()
df['price_m2'] = df['price_m2'].astype('int')

#prevent SettingWithCopyWarning message from appearing
pd.set_option('chained_assignment', None)

# День недели публикации объявления (0 — понедельник, 1 — вторник и так далее).
df['day'] = pd.DatetimeIndex(df['first_day_exposition']).weekday

# Месяц публикации объявления.
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month

# Год публикации объявления.
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year

# Тип этажа квартиры (значения — «первый», «последний», «другой»). Для этого необходимо написать функцию.
def type_floor(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
df['type_floor'] = df.apply(type_floor, axis=1)

# Расстояние до центра города в километрах (переведите из м в км и округлите до целых значений).
df['city_centers_nearest'] = df['city_centers_nearest'] / 1000
df['city_centers_nearest'] = df['city_centers_nearest'].round()
df['city_centers_nearest'] = df['city_centers_nearest'].astype(pd.Int32Dtype())
```

**Срок продажи квартиры**
    
По результату узучения данных с количеством дней размещения объявлений видно, что чаще всего квартира продаётся за 50 дней, среднее значение - 177 дней и медиана - 93 дня. Быстрой продажу можно назвать срок до 44 дней. А необычайно долгой более 225 дней.
    
**Факторы, влияющие на стоимость**
- Самые дешевые квартиры находятся на первом и последнем этажах. Наибольшая стоимость у тех, которые расположены посередине.
- Отчетлива видная зависимость цены от общей площади (0,8), жилой площади(0,7), площади кухни(0,6), количества комнат (0,5) и от высоты потолков (0,4).
- Зависимость цены от даты размещения (день недели, месяц, год) не видна, но можно выделить, что со вторника по четверг и в августе размещались самые дорогие квартиры, а в субботу и в июне, самые дешевые. Ещё стоит обратить вниимание, что до 2018 года квартиры дешевели, а после 2019 начался подъем в цене.
